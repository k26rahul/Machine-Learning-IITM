```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_candidates = np.linspace(-1.5, 1.5, 400)
y_candidates = np.linspace(-1.5, 1.5, 400)

curve_x = []
curve_y = []

for x in x_candidates:
  for y in y_candidates:
    lhs = (x**2 + y**2 - 1)**3 - x**2 * y**3
    if abs(lhs) < 10**-3:
      curve_x.append(x)
      curve_y.append(y)

len(curve_x), len(curve_y)
plt.scatter(curve_x, curve_y, color='red')
plt.grid()
```

```python
def heart_equation(x, y):
  return (x**2 + y**2 - 1)**3 - x**2 * y**3


x = np.linspace(-1.5, 1.5, 400)
y = np.linspace(-1.5, 1.5, 400)

X, Y = np.meshgrid(x, y)
Z = heart_equation(X, Y)

plt.contour(X, Y, Z, levels=[0], colors='red')
plt.grid()
```
