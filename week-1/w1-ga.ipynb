```python
import numpy as np
```

```python
data = np.array([
    [1, 0.5],
    [2, 2.3],
    [3, 3.1],
    [4, 3.9],
])


def encode1(x):  # x is vector
  return x[0] - x[1]


def decode1(u):  # u is scalar
  return np.array([u/2, u/2])


loss = 0
for x in data:
  u = encode1(x)
  x_decoded = decode1(u)
  loss += sum((x - x_decoded) ** 2)

loss/4
```

```python
import matplotlib.pyplot as plt
```

```python
data = np.array([
    [-1, 0.0319],
    [0, 0.8692],
    [1, 1.9566],
    [2, 3.0343],
])

x = data[:, 0]
y = data[:, 1]

_ = plt.scatter(x, y)

params = [
    [1, 1, 'r'], [1, 2, 'g'], [2, 1, 'b'], [2, 2, 'y']
]
for a, b, color in params:
  y_pred = x * a + b
  mse = sum((y - y_pred) ** 2)/4
  print(a, b, mse)
  _ = plt.plot(x, y_pred, color=color)
```
