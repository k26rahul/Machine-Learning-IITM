```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_xy(x, y):
  plt.figure(figsize=(6, 4))
  plt.plot(x, y)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.grid(True)
  plt.axhline(0, color='gray', lw=2)
  plt.axvline(0, color='gray', lw=2)
```

```python
x = np.linspace(-10, 10, 500)
y = x
plot_xy(x, y)
```

```python
x = np.linspace(-10, 10, 500)
y = x**2
plot_xy(x, y)
```

```python
x = np.linspace(-10, 10, 500)
y = x**3
plot_xy(x, y)
```

```python
x = np.linspace(-10, 10, 500)
y = x**20
plot_xy(x, y)
```

```python
x = np.linspace(-10, 10, 500)
y = x**21
plot_xy(x, y)
```

```python
def plot_xy_trig(x, y, xlim=2):
  plot_xy(x, y)
  xticks = np.pi * np.arange(-xlim, xlim+1, 1)
  xtick_labels = [f'{i}' if i != 0 else '0' for i in range(-xlim, xlim+1)]
  plt.xticks(xticks, xtick_labels)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y = np.sin(x)
plot_xy_trig(x, y)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y = np.cos(x)
plot_xy_trig(x, y)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y = np.sin(x) + np.cos(x)
plot_xy_trig(x, y)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y = np.sin(x)**2 + np.cos(x)**2
plot_xy_trig(x, y)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y = np.sin(x**2)
plot_xy_trig(x, y)
```

```python
def plot_contour(X, Y, Z, levels=20):
  plt.contour(X, Y, Z, levels=levels)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.colorbar(label='z')
  plt.axis('equal')
  plt.grid(True)


def plotly_surface(X, Y, Z):
  import plotly.graph_objects as go
  fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='Viridis')])
  fig.update_layout(
      width=600,
      height=600,
      scene_camera=dict(
          eye=dict(x=0, y=2, z=0),
          up=dict(x=0, y=0, z=1),
          center=dict(x=0, y=0, z=0)
      ),
      scene=dict(aspectmode='cube')
  )
  return fig
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X

plot_contour(X, Y, Z)
plotly_surface(X, Y, Z)
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X + Y

plot_contour(X, Y, Z)
plotly_surface(X, Y, Z)
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2

plot_contour(X, Y, Z)
plotly_surface(X, Y, Z)
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plot_contour(X, Y, Z)
plotly_surface(X, Y, Z)
```
